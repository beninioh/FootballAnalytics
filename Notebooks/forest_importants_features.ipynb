
### Determinant features to win a football games.
###### The investigation is based on the 5 top leagues fron 2016 to 2020.
###### It is obvious that goals or assist are determinants to win so we do not consider those attributes or any'%_' added attributes.


```python
import matplotlib.pyplot as plt
from analysis.random_forest import *
import seaborn as sns

_, r, f = train_forest(['ligue1', 'prleague', 'liga'],
                       ['1617', '1718', '1819', '1920'],
                       db='../rotowire/database/GamesInfo.db',
                       interpret=lambda x: interpret_prediction3(x))

_, r_l1, f_l1 = train_forest(['ligue1'],
                             ['1617', '1718', '1819', '1920'],
                             db='../rotowire/database/GamesInfo.db',
                             interpret=lambda x: interpret_prediction3(x)
                             )

_, r_lga, f_lga = train_forest(['liga'],
                               ['1617', '1718', '1819', '1920'],
                               db='../rotowire/database/GamesInfo.db',
                               interpret=lambda x: interpret_prediction3(x))

_, r_prl, f_prl = train_forest(['prleague'],
                               ['1617', '1718', '1819', '1920'],
                               db='../rotowire/database/GamesInfo.db',
                               interpret=lambda x: interpret_prediction3(x))
```

```python
print('Global accuracy : ', r * 100, ' %')
print('Ligue 1 accuracy : ', r_l1 * 100, ' %')
print('Liga accuracy : ', r_lga * 100, ' %')
print('Pr league accuracy : ', r_prl * 100, ' %')
```

```python
g = sns.barplot(x=f.index,
                y=f.importances,
                palette=sns.color_palette("GnBu_d"))
g.set_xticklabels(g.get_xticklabels(), rotation=90)
plt.title('Important features to win.')
plt.show()

g = sns.barplot(x=f_l1.index,
                y=f_l1.importances,
                palette=sns.color_palette("GnBu_d"))
g.set_xticklabels(g.get_xticklabels(), rotation=90)
plt.title('Important features to win in Ligue1.')
plt.show()

g = sns.barplot(x=f_lga.index,
                y=f_lga.importances,
                palette=sns.color_palette("GnBu_d"))
g.set_xticklabels(g.get_xticklabels(), rotation=90)
plt.title('Important features to win in Liga.')
plt.show()

g = sns.barplot(x=f_prl.index,
                y=f_prl.importances,
                palette=sns.color_palette("GnBu_d"))
g.set_xticklabels(g.get_xticklabels(), rotation=90)
plt.title('Important features to win in pr league.')
plt.show()
```
