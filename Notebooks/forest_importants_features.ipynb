
### Determinant features to win a football games.
###### The investigation is based on the 5 top leagues fron 2016 to 2020.
###### It is obvious that goals or assist are determinants to win so we do not consider those attributes or any'%_' added attributes.


```python
import matplotlib.pyplot as plt
from random_forest import *
import seaborn as sns

_, r, f = train_forest(['ligue1', 'prleague'],
                       ['1617', '1718', '1819', '1920'],
                       db='../rotowire/database/GamesInfo.db')
```

```python
print('Accuracy : ', r * 100, ' %')
f = f.query('importances > 0.01')
f
```

```python
g = sns.barplot(x=f.index,
                y=f.importances,
                palette=sns.color_palette("GnBu_d"))
g.set_xticklabels(g.get_xticklabels(), rotation=90)
plt.title('Important features to win.')
plt.show()
```
